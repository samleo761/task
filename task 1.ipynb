```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests as rq
```

```python
url="https://www.flipkart.com/search?q=rog+gaming+phone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_8_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_2_8_na_na_na&as-pos=2&as-type=RECENT&suggestionId=rog+gaming+phone&requestId=f6fd4819-1414-4c60-9d87-016ad079e21c&as-searchtext=rog%20gami";
request = rq.get(url)
request
```

```python
html = bs(request.text)
html
```

```python
mobile_name = html.select("._4rR01T")
storage = html.select(".rgWa7D:nth-child(1)")
specs_and_features = html.select(".rgWa7D:nth-child(5)")
price = html.select("._1_WHN1")
```

```python
details = pd.DataFrame(mobile_name,columns=["mobile name"])
details["storage"] = storage
details["processor"] = specs_and_features
details["price"] = price
details
```

```python
details.to_csv("rog phones.csv")
```
